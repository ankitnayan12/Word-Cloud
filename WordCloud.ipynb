```
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd
```

```
file = open('/content/drive/My Drive/Colab Notebooks/detecting cancer using comuter vision/bhanot2016.pdf.txt')
```

```
lines = [line.strip() for line in file.readlines()]

speech = " ".join(lines)
```

```
speech
```

```
from nltk.corpus import stopwords
```

```
wc_bhanot2016_pdf = WordCloud(width = 800, height = 800, background_color ='white', stopwords = STOPWORDS)
```

```
wc_bhanot2016_pdf.generate(speech)
```

```
plt.figure(figsize = (8, 8), facecolor = None) 
plt.title("wordcloud", fontsize=30, pad=30)
plt.imshow(wc_bhanot2016_pdf) 
plt.axis("off") 
plt.tight_layout(pad = 0)
# plt.savefig('bhanot2016_pdf_word_cloud.pdf')
plt.show()
```
